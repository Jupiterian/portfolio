```python
%%%html

<div id="outputPH2">Results will appear here:<br></div>

<script>

(() => {

    const output = document.getElementById("outputPH2");

    let snackChoice = prompt("Choose a number: 1, 2, or 3 to get a snack!");

    if (snackChoice == "1") { // Chocolate bar
        let chocolateType = prompt("You got a chocolate bar! Choose your type: milk or dark?");
        
        // Nested condition for chocolate type
        if (chocolateType.toLowerCase() === "milk") { 
            output.innerText += "You chose milk chocolate \n";
        } else if (chocolateType.toLowerCase() === "dark") { 
            output.innerText += "You chose dark chocolate  (fancy!)\n";
        } else {
            output.innerText += "Thatâ€™s not a chocolate type I recognize! \n";
        }

    } else if (snackChoice == "2") { // Cookies
        output.innerText += "You got a cookie! Sweet and delicious \n";

    } else if (snackChoice == "3") { // Candy
        let candyColor = prompt("You got candy! Choose your color: red, green, or yellow?");
        
        // Nested condition for candy color
        if (candyColor.toLowerCase() === "red") { 
            output.innerText += "You chose red candy! \n";
        } else if (candyColor.toLowerCase() === "green") { 
            output.innerText += "You chose green candy! \n";
        } else if (candyColor.toLowerCase() === "yellow") { 
            output.innerText += "You chose yellow candy! \n";
        } else {
            output.innerText += "That is a currently unknown candy color.! \n";
        }

    } else {
        output.innerText += "Invalid choice! The vending machine is not skilled. \n";
    }

    // Challenge: 10% chance of extra cookie
    if (Math.random() < 0.1) { // Random number between 0 and 1 â€” 10% chance
        output.innerText += "Bonus treat! The machine gives you an extra cookie! \n";
    }

})();
</script>
```
